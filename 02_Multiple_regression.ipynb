# **Linear Regression (Multiple Regression)**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/homeprices.csv')
df
```

```python
# Data preprocessing
import math
# Taking median to fill the missing values
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
# filling the missing value in the dataset
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
df
```

```python
# creating linear regression class object
reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms', 'age']], df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```

```python
# coefficent*x1 + coefficent*bedrooms + coefficient*age + intercept
112.06244194*3000 + 23388.88007794*3 +(-3231.71790863)*40 + 221323.00186540396
```

```python
reg.predict([[2500, 4, 5]])
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from word2number import w2n
```

```python
df = pd.read_csv('/content/tableConvert.com_s9yhrm.csv')
df
```

```python
# Renaming column name
new_column = ['experience', 'test_score', 'interview_score', 'salary']
df.columns = new_column
df
```

```python
df.experience = df.experience.fillna("zero")
df
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
```

```python
df
```

```python
meadian_test_score = math.floor(df['test_score'].mean())
meadian_test_score
```

```python
df['test_score'] = df['test_score'].fillna(meadian_test_score)
print(df)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience', 'test_score', 'interview_score']], df.salary);
```

```python
reg.predict([[2, 9, 6]])
```
