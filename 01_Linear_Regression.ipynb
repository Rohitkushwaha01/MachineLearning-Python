### **Machine Learning: Linear Regression with One variable**

```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#loading data

df = pd.read_csv("/content/linearRegressionDataset.csv");
df
```

```python
%matplotlib inline
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.Area, df.Price, color='red', marker='+')
```

```python
df.Area
```

```python
# dropping the price column
new_df = df.drop('Price', axis='columns')
# new_df = df.Area this won't work
new_df
```

```python
price = df.Price
price
```

```python
# linear regression
# create linear regression object
reg = linear_model.LinearRegression()

# fit the data: Fitting the data means you are training the model
reg.fit(new_df,price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# y = mx + b
m = 135.78767123 # coefficent
x = 3300 # area which we want to predict
b = 180616.43835616432 # reg.intercept_
y = m*x+b
y
```

 **Predictions**

```python
area_df = pd.read_csv('/content/Area.csv')
area_df.head(5)
```

```python
predicted_price = reg.predict(area_df)
predicted_price
```

```python
area_df['price'] = predicted_price
area_df
```

```python
area_df.to_csv('prediction_output.csv')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/canada_per_capita_income.csv")
df.head(5)
```

```python
#changing column names
new_column = ["year", "income"]
df.columns = new_column
df.head(3)
```

```python
plt.xlabel("year")
plt.ylabel("income")
plt.scatter(df.year, df.income, color='red');
```

```python
Year = df.drop('income', axis='columns')
Year.head(3)
```

```python
Income = df.income
Income.head(3)
```

```python
#creating model
reg = linear_model.LinearRegression()

reg.fit(Year, Income)
```

```python
reg.predict([[2020]])
```
